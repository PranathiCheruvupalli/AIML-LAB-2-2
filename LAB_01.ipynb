part-02: visualize

```python
people=['ann','brandon','chen','david','emily','farook']
age=[21,22,32,31,28,25]
weight=[55,58,63,59,50,49]
height=[160,135,170,165,173,168]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(weight,height)
plt.title("relationship between weight and height of patients")
plt.ylabel("height in meters")
plt.xlabel("weight of patients in kgs")
plt.show()
```

```python
plt.scatter(age,weight)
plt.title("relationship between age and weight of patients")
plt.ylabel("weight of patients in kgs")
plt.xlabel("age of patients")
plt.show()
```

```python
plt.scatter(age,height)
plt.title("relationship between age and height")
plt.ylabel("height of patients in meters")
plt.xlabel("age of patients")
plt.show()
```

generate bar plot

```python
plt.figure(figsize=(12,4))
plt.title("people's weight in kgs",fontsize=16,fontstyle='italic')
#main plot function
plt.bar(x=people,height=weight,width=0.6,color='orange',edgecolor='k',alpha=0.6)
plt.xlabel("people",fontsize=15)
#plt.xticks(fontsize=14,rotation=30)
#plt,yticks(fontsize=14)
plt.ylabel("weight (in kgs)",fontsize=15)
plt.show()
```

generate histogram

```python
import numpy as np
plt.figure(figsize=(7,5))
#main plot function 'hist'
plt.hist(weight,color='red',edgecolor='k',alpha=0.75,bins=5)
plt.title("histogram of patient weight",fontsize=18)
plt.xlabel("weight in kgs",fontsize=15)
plt.show()
```
